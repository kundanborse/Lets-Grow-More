```python
img = cv2.imread(r'C:\Users\Nishant\Documents\kundan borse\my photos\cat.jpeg')  # in input_image give path of your image
```

```python

# Convert the image to grayscale
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

inverted_image = 255 - gray_image

blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)

pencil_sketch = cv2.divide(gray_image, 255 - blurred, scale=256)

# Save the pencil sketch image
cv2.imwrite('output_image.jpg', pencil_sketch)
```
